### 関数とグラフの作成

```python
from sympy import init_printing, symbols, plot
%matplotlib inline
```

```python
# SymPyを使った関数
init_printing(use_latex="mathjax")

x = symbols("x")
```

```python
f = x ** 2
f
```

```python
# fにx = 2を代入
f.subs(x, 2)
```

```python
### 関数のプロット
# f = x**2 のプロット
plot(f, (x, -2, 2), legend=True)
```

### いろいろな関数とグラフ

```python
### 多項式関数
plot(x + 1, (x, -2, 2), legend=True)
```

```python
### 複雑な多項式
plot(x**3 + 2*x**2 - 19*x, (x, -6, 5), legend=True)
```

```python
### 三角関数
# 半径=1の円は単位円と呼ばれる
# 媒介変数表示：2つの変数（x, y）を共通の変数（ここではΘ）で表すこと
from sympy import sin, cos, tan, pi
from sympy.plotting import plot_parametric

# 描画サイズを5x5の正方形に指定する
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = 5, 5
```

```python
# 媒介変数tはthetaの頭文字
t = symbols("t")

# 媒介変数表示を使用した単位円のプロットを作成
plot_parametric(cos(t), sin(t), (t, 0, 2*pi), legend=True)

# 描画サイズをdefault値に戻す
plt.rcParams["figure.figsize"] = 6, 4
```

```python
# sinxとcosxのプロット
# show=Falseとして、この行ではグラフは表示させない
p = plot(sin(x), cos(x), (x, -2*pi, 2+pi), legend=True, show=False)

# グラフの色を指定
p[0].line_color = "b" # 青
p[1].line_color = "r" # 赤

# グラフの表示
p.show()
```

```python
# cos(x)をx方向に+pi/2平行移動
cos(x-pi/2)
```

```python
### 奇関数と偶関数
# 奇関数：f(-x) = -f(x) 原点に対して対称な形を取る
# sinxは奇関数
sin(-x)
```

```python
# 偶関数：y軸に対して対称な形をとる
# cosxは偶関数
cos(-x)
```

```python
# tanxのプロット
p = plot(tan(x), (x, 0, 2*pi), legend=True, show=False)

# y軸の表示範囲を指定
p.ylim = (-10, 10)

p.show()
```

### 指数関数

```python
### 指数関数のプロット
p = plot(2**x, 1/2**x, (x, -2, 2), legend=True, show=False)
p[0].line_color = "b"
p[1].line_color = "r"

p.show()
```

### 対数関数

```python
from sympy import log

# log(変数, 底)
log(4, 2)
```

```python
# 対数関数のプロット
p = plot(log(x, 2), log(2*x, 2), (x, 0.1, 100),
        legend=True, show=False)
p[0].line_color = "b"
p[1].line_color = "r"

p.show()
```

```python
### 指数関数と対数関数のプロット
p = plot(2**x, log(x, 2), x, (x, -2, 4), legend=True, show=False)
p.ylim = (-2, 4)

p[0].line_color = "b"
p[1].line_color = "r"
p[2].line_color = "g"

p.show()
```
