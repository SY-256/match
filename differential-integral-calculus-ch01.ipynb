```python
import numpy
import pandas as pd
import math

from sympy import init_printing, symbols, factor, expand, solveset, linsolve
```

### 第1章：SymPyの基礎

```python
# 数式の出力をLaTexで表記する
init_printing(use_latex="mathjax")

# xを記号を定義する
x = symbols("x")

x + x - 1
```

```python
### 文字式への代入
x,y = symbols("x y")
f = x**2 - y**2

f
```

```python
# x, yをキーにした辞書で値の指定ができる
f.subs([(x, 1), (y, 2)])
```

```python
# 文字の代入もできる
a = symbols("a")

# x=aとy=2の代入
f.subs([(x, a), (y, 2)])
```

```python
### 式の因数分解と展開(factor, expand)

# factorを使用して、f = x**2 - y**2を因数分解
f_factor = factor(f)
f_factor
```

```python
# expandを使用して、(x - y)(x + y)を展開
expand(f_factor)
```

```python
### 方程式を解く
# solvesetを使用して方程式を解く
# 右辺を0にする必要がある
solveset(x + 10 - 8)
```

```python
### 2次方程式
b, c = symbols("b c")
solveset(a*x**2 + b*x + c, x)
```

```python
### 連立方程式

# 連立方程式の入力
eq1 = x + y - 5
eq2 = x - 2*y + 19

# lisolveを使用して連立方程式を解く
linsolve([eq1, eq2], (x, y))
```
